```python
import requests
from bs4 import BeautifulSoup
import time
from selenium.webdriver import Chrome
import pandas as pd
driver = Chrome(executable_path='/Users/matt/Code/SpaceScraper/chromedriver')
```

## Get SpaceX listings

```python
driver.get('https://www.spacex.com/careers/?department=')
time.sleep(3)
aTags = driver.find_elements_by_tag_name('a')
time.sleep(3)
print(len(aTags), "a tags found")

def filter_spacex(x):
    x = x.lower()
    cond_a = "software" in x
    cond_b = "data" in x
    cond_c = "sr" not in x
    cond_d = "internship" not in x
    if ((cond_a == True) | (cond_b == True)) & (cond_c == True) & (cond_d == True):
        return True
    else: 
        return False
        
spacex_jobs_df = pd.DataFrame([(e, i.text, i.get_property("href"))  for e, i in enumerate(aTags) if filter_spacex(i.text) == True])
spacex_jobs_df.columns = ["idx", "title", "link"]

# Exclude jobs I've already applied for 
applied = [
    'https://boards.greenhouse.io/spacex/jobs/5447125002?gh_jid=5447125002',
    'https://boards.greenhouse.io/spacex/jobs/5947405002?gh_jid=5947405002'
]

spacex_jobs_df = spacex_jobs_df[spacex_jobs_df['link'].apply(lambda x: x not in applied)]
```

## Get Firefly listings

```python
firefly_board = 'https://firefly.hrmdirect.com/employment/job-openings.php?search=true&'
driver.get(firefly_board)
time.sleep(3)
aTags = driver.find_elements_by_tag_name('a')
time.sleep(3)

def filter_firefly(tag):
    text = tag.text.lower()
    link = tag.get_property("href")

    cond_a = "hrmdirect" in link
    cond_b = "software" in text
    if (cond_a == True) & (cond_b == True):
        return True
    else: 
        return False

firefly_jobs_df = pd.DataFrame([(e, i.text, i.get_property("href"))  for e, i in enumerate(aTags) if filter_firefly(i) == True])
```
